```python
import numpy as np    # linear algebra
import pandas as pd   # data processing, CSV file I/O
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)

# We don't need this data
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(1)
```

```python
import ast
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head(1)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(1)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head(1)
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head(1)
```

```python
# convert in lower case
new['tags'] = new['tags'].apply(lambda x:x.lower())
```

```python
new.head(1)
```

#Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

# Stemming


```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
new['tags'][0]
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
# Recommend movie name
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```

```python
from IPython.display import display, HTML
import ipywidgets as widgets
from IPython.display import clear_output

# Function to recommend movies
def recommend(movie):
    # Find the index of the movie in the dataset
    index = new[new['title'] == movie].index[0]

    # Calculate similarity scores and sort them
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

    # Get the top 5 recommendations (excluding the first one which is the movie itself)
    recommended_movies = [new.iloc[i[0]].title for i in distances[1:6]]

    # Create a DataFrame with serial numbers
    recommendation_df = pd.DataFrame(recommended_movies, columns=['Recommended Movies'])
    recommendation_df.index += 1  # Start index from 1

    # Display the recommendations as a table
    display(HTML(recommendation_df.to_html(index=True, header=True)))

# Function to handle button click
def on_button_click(b):
    clear_output(wait=True)  # Clear previous output but keep the widgets visible

    # Display the input box and button again
    display(movie_input, button)

    movie = movie_input.value
    if movie in new['title'].values:
        recommend(movie)
    else:
        display(HTML(f"<h3>Movie '{movie}' not found in the dataset.</h3>"))

# Create a text input box
movie_input = widgets.Text(
    value='Gandhi',
    placeholder='Enter movie title',
    description='Movie:',
    disabled=False
)

# Create a button
button = widgets.Button(
    description='Recommend',
    disabled=False,
    button_style='',
    tooltip='Click to get recommendations',
    icon='check',
    bg_color='lightgreen',
)

# Set up the button click event
button.on_click(on_button_click)

# Display the input box and button initially
display(movie_input, button)
```
